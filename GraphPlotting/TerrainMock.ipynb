```python
import math
import numpy as np

def GetHeight(xGrid, yGrid):
    return np.cos(xGrid/10+3.14/2) * np.sin(yGrid/10) + 1

stencil = np.arange(0, 101, 1)
Y, X = np.meshgrid(stencil, stencil)
Z = GetHeight(X, Y)
```

```python
import matplotlib.pyplot as plt

h = plt.contourf(X, Y, Z)
plt.axis('scaled')
plt.colorbar()
plt.show()
```
